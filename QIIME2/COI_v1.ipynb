First, import the paired forward and reverse sequences with a manifest file. Manifest prepared by listing all files in bash and formatting in Excel. 

```python
#Import the fastq files into a QIIME2 artifact 

!qiime tools import \
  --type 'SampleData[PairedEndSequencesWithQuality]' \
  --input-path leray_manifest_runbb_v2_grid.csv \
  --output-path COI_1-paired-end-demux.qza \
  --input-format PairedEndFastqManifestPhred33
```

Trim reverse primers
See: https://docs.qiime2.org/2020.6/plugins/available/cutadapt/trim-paired/
default is to match IUPAC wildcards ( --p-match-adapter-wildcards T )
- these should be the F/R marker primers
- cannot handle inosine bases; using N instead
- f primer/ r primer
- or could just trim off length of primer+barcode? this provides a test of primer accuracy/error 


```python
!qiime cutadapt trim-paired \
    --i-demultiplexed-sequences COI_1-paired-end-demux.qza \
    --p-front-f '^GGWACWGGWTGAACWGTWTAYCCYCC' \
    --p-front-r '^TANACYTCNGGRTGNCCRAARAAYCA' \
    --o-trimmed-sequences COI_1-demux-pairedend-trimmed1.qza \
    --verbose
```

```python
!qiime demux summarize \
  --i-data COI_1-demux-pairedend-trimmed1.qza \
  --o-visualization COI_1-demux-pairedend-trimmed1.qzv
```

```python
q2.Visualization.load('COI_1-demux-pairedend-trimmed1.qzv')
```

```python
#Generate ESVs (Exact Sequence Variants). QIIME2 has two packages available for this. We are using dada2 for our single end reads.
# NOTE: if you are wanting to do paired end reads, you'll use qiime dada2 denoise-paired (https://docs.qiime2.org/2019.7/plugins/available/dada2/denoise-paired/)
# NOTE: if you are wanting to join reads you'll have to use Deblur


!qiime dada2 denoise-paired \
  --i-demultiplexed-seqs COI_1-demux-pairedend-trimmed1.qza \
  --p-trunc-len-f 258 \
  --p-trunc-len-r 168 \
  --p-pooling-method pseudo \
  --p-chimera-method pooled \
  --p-n-threads 0 \
  --p-n-reads-learn 10000 \
  --o-representative-sequences COI_1-paired-end-demux_rep-seqs-dada2.qza \
  --o-table COI_1-paired-end-demux_table-dada2.qza \
  --o-denoising-stats COI_1-paired-end-demux_stats-dada2.qza
```

```python
# The feature-table summarize command will give you information on how many sequences are associated with each sample and with each feature, histograms of those distributions, and some related summary statistics. 
!qiime feature-table summarize \
  --i-table COI_1-paired-end-demux_table-dada2.qza \
  --o-visualization COI-table.qzv \
  
#--m-sample-metadata-file metadata_COI_clean.txt
```

```python
##filter out detected chimeras

!qiime feature-table filter-features \
  --i-table COI-paired-end-demux_table-dada2_vsearch97.qza \
  --m-metadata-file uchime-dn-out/nonchimeras.qza \
  --o-filtered-table uchime-dn-out/table-nonchimeric-wo-borderline.qza
!qiime feature-table filter-seqs \
  --i-data COI-paired-end-demux_rep-seqs-dada2_vsearch97.qza \
  --m-metadata-file uchime-dn-out/nonchimeras.qza \
  --o-filtered-data uchime-dn-out/rep-seqs-nonchimeric-wo-borderline.qza
!qiime feature-table summarize \
  --i-table uchime-dn-out/table-nonchimeric-wo-borderline.qza \
  --o-visualization uchime-dn-out/table-nonchimeric-wo-borderline.qzv
```
